```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df
```

Seperating numerical and categorical values from the number column

```python
df['number_numerical']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
df.head()
```

```python
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

Splitting the ctegorical and numbers from the column Cabin

```python
df['Seat_number']=df['Cabin'].str.extract('(\d+)')
```

```python
df.head()
```

```python
df['Cabin_Number']=df['Cabin'].str.extract('(\D+)')
```

```python
df.head()
```

```python
df["Ticket_cat"]=df['Ticket'].apply(lambda x:x.split()[0])
```

```python
df['Ticket_cat']
```

```python
df['Ticket_cat']=np.where(df['Ticket_cat'].str.isdigit(),np.nan,df['Ticket_cat'])
```

```python
df['Ticket_cat']
```

```python
df['Ticket_number']=df['Ticket'].apply(lambda x:x.split()[-1])
```

```python
df['Ticket_number']
```

```python
df['Ticket_number']=pd.to_numeric(df['Ticket_number'],errors='coerce',downcast='integer')
```
